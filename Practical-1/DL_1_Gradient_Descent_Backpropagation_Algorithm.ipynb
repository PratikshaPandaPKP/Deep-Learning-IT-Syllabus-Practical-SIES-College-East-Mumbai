```python
import numpy as np

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


# Given data
x1 = 0.8  # Input
y = 0.95  # Target output
w1 = 0.5  # Initial weight from input to hidden
w2 = 0.6  # Initial weight from hidden to output
eta = 0.1  # Increased learning rate
iterations = 100  # Number of iterations

# Loop for multiple iterations
for epoch in range(1, iterations + 1):
    # Forward propagation
    zh = x1 * w1  # Net input to hidden layer
    h = sigmoid(zh)  # Activation of hidden layer

    zo = h * w2  # Net input to output layer
    o = sigmoid(zo)  # Activation of output layer

    # Error calculation
    error = 0.5 * (y - o)**2

    # Backward propagation
    # Delta for output layer
    delta_o = (y - o) * o * (1 - o)

    # Update weight from hidden to output layer
    dw2 = eta * delta_o * h
    w2_new = w2 + dw2

    # Delta for hidden layer
    delta_h = delta_o * w2 * h * (1 - h)

    # Update weight from input to hidden layer
    dw1 = eta * delta_h * x1
    w1_new = w1 + dw1

    # Print the iteration number and updated weights at key intervals
    if epoch % 10 == 0 or epoch == 1:
        print(f"Iteration {epoch}:")
        print(f"  Updated weight w1: {w1_new}")
        print(f"  Updated weight w2: {w2_new}")
        print(f"  Error: {error}\n")

    # Update weights for next iteration
    w1, w2 = w1_new, w2_new
```
