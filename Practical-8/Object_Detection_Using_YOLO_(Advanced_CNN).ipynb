```python
import torch
import cv2

# Load the YOLOv5 model directly using torch.hub
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/yolov5/yolov5s.pt')  # Ensure correct path to yolov5s.pt

# Load image
image = cv2.imread('/content/Object.png')

# Run inference
results = model(image)

# Show results
results.show()  # This will display the image with detected objects
```
